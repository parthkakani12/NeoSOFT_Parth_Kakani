```python
outlier_summary = {}

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    outlier_summary[col] = len(outliers)

outlier_df = (
    pd.DataFrame(
        outlier_summary.items(),
        columns=["Column", "Outlier_Count"]
    )
    .sort_values(
        by="Outlier_Count",
        ascending=False
    )
)

outlier_df
```

## Categorical Feature Analysis

For categorical feature analysis, all categorical columns were grouped together using Pandas. EDA was then performed to understand the distribution and characteristics of categorical variables. The analysis included missing value detection, unique category analysis, frequency distribution, percentage distribution, and visualization using bar charts. This helped identify dominant categories, data imbalance, and patterns within categorical features.


```python
corr_matrix = df[num_cols].corr().abs()

high_corr = (
    corr_matrix
    .where(corr_matrix > 0.7)
    .stack()
    .reset_index()
)

high_corr.columns = ["Feature_1", "Feature_2", "Correlation"]

high_corr = high_corr[
    high_corr["Feature_1"] != high_corr["Feature_2"]
]

high_corr.drop_duplicates(
    subset=["Correlation"],
    inplace=True
)

high_corr.sort_values(
    by="Correlation",
    ascending=False
)
```
