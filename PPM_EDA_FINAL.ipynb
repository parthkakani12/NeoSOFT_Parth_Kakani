```python
outlier_summary = {}

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    outlier_summary[col] = len(outliers)

outlier_df = (
    pd.DataFrame(
        outlier_summary.items(),
        columns=["Column", "Outlier_Count"]
    )
    .sort_values(
        by="Outlier_Count",
        ascending=False
    )
)

outlier_df
```

```python
missing_percentage = (
    df[num_cols]
    .isnull()
    .mean()
    * 100
)

missing_percentage.sort_values(
    ascending=False
)
```

Credit Bureau Missing Value Analysis

```python
bureau_cols = [
    "Credit_Risk",
    "CreditScore1_CreditScore",
    "CreditScore1_CreditScoreCategory",
    "StrikeDates1_StrikeDate1",
    "StrikeDates1_StrikeDate2",
    "StrikeDates1_StrikeDate3",
    "Judgements1_PersonName",
    "Judgements1_Plaintiff",
    "Judgements1_Attorneys",
    "Judgements1_JudgmentAmount",
    "EagleEye1_UnderDebtReview"
]

bureau_summary = pd.DataFrame({
    "Missing_Count": df[bureau_cols].isnull().sum(),
    "Available_Count": df[bureau_cols].notnull().sum(),
    "Missing_Percentage": (
        df[bureau_cols].isnull().mean() * 100
    ).round(2)
})

bureau_summary.sort_values(
    by="Missing_Percentage",
    ascending=False
)
```

Leakage Analysis

Current_Payment is the target variable.

Features such as:

Current_Payment_Amount
Current_Cure

are generated during the same billing cycle as the target and therefore cannot be used as predictors.

These columns will be excluded from model training to avoid target leakage.
